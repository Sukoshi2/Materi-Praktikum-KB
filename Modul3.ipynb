```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_categorial = pd.read_csv('dataset/House_Rent_Dataset.csv')
dataset_categorial.head(2)
```

```python
dataset_timeseries = pd.read_csv('dataset/DailyDelhiClimateTest.csv')
dataset_timeseries.head(2)
```

```python
x = dataset_categorial['City']
y = dataset_categorial['Rent']
```

```python
#plt.bar(dataset_categorial['City'], dataset_categorial['Rent'])
plt.bar(x,y,
       align='edge', width=1)
plt.title('Nama Kota', loc='right')
plt.xlabel('Kota', loc='right')
plt.ylabel('Sewa', loc='bottom')
plt.show()
```

```python
plt.plot(dataset_timeseries['date'],
         dataset_timeseries['humidity'],
         'r',
         linewidth=2,
         linestyle='dashed',
         marker='o', markersize=3)
plt.plot(dataset_timeseries['date'], dataset_timeseries['wind_speed'],'b')
plt.title('Time Series Data', loc='right')
plt.xlabel('Date', loc='right')
plt.ylabel('Angka', loc='bottom')
plt.legend(['kelembapan','kecepatan angin'])
plt.show()
```

```python
df = dataset_categorial.dropna(subset=['City'])
```

```python
data_pie = df['City'].value_counts().rename_axis('City').reset_index(name='id_count')
```

```python
plt.figure(figsize=(10,10))
plt.pie(data_pie.id_count, labels=data_pie.City, startangle=90,
       autopct='%.1f%%')
plt.title('Nyoba')
plt.show()
```

```python
sns.scatterplot(data=dataset_categorial, x='City', y='Rent')
sns.scatterplot(data=dataset_categorial, x='City', y='Size')
plt.show()
```

```python
#used = dataset_timeseries.isnull().values.any()
sns.boxplot(dataset_timeseries['wind_speed'])
plt.show()
```

```python
sns.heatmap(dataset_timeseries.corr(), cmap='Greens')
plt.show()
```
