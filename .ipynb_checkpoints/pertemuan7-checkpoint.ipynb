### Slicing dan indexing

```python
var_random = tf.random.normal([16,16,32,64])
var_random
```

```python
var_random[2:3]
```

```python
var_random[1][2][1]
```

```python
tf.gather(var_random, axis=0, indices=[0,2,1,1])
```

```python
tf.gather_nd(var_random, indices= [[1,1],[2,2]])
```

```python
var_random.get_shape()
```

```python
tf.reshape(var_random, (64, 16, 32, 16)) # harus sama jumlah isinya dengan cara dikali = dengan shape awal
```

```python
print(var_random.shape)
var_random1 = tf.reshape(var_random, (64,16,32,16))
print(var_random1.shape)
```

### Expand Dimension

```python
print(var_random.shape)
var_random1 = tf.expand_dims(var_random, axis=1)
print(var_random1.shape)
```

### Squeezeing

```python
print(var_random1.shape)
var_random1 = tf.squeeze(var_random1, axis=3)
print(var_random1.shape)
```

### Operator Aritmatika

```python
a = tf.constant([[2,3], [4,5]])
b = tf.constant([[9,8], [6,7]])
```

```python
tf.add(a,b)
```

```python
tf.subtract(a,b)
```

```python
tf.multiply(a,b)
```

```python
tf.divide(a,b)
```

```python
tf.math.log(a,b)
```

```python
tf.pow(a,b)
```

```python
tf.matmul(a,b)
```
