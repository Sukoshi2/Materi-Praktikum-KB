### Hierarchical Clustering

```python
data_iris = skdt.load_iris()
```

```python
df_iris = pd.DataFrame(data_iris.data, columns=data_iris.feature_names)
df_iris.rename(columns = {'sepal length (cm)': 'sl', 'sepal width (cm)' : 'sw', 
                          'petal length (cm)' : 'pl', 'petal width (cm)' : 'pw'}, inplace=True)
df_iris
```

```python
points_sepal = df_iris.iloc[:,[0,1]].values
points_petal = df_iris.iloc[:,[2,3]].values
```

```python
hr_linkage = linkage(points_sepal, method='ward')
hr_complete = complete(points_petal)
```

```python
plt.figure(1, figsize=(14,7))
dendrogram(hr_linkage)
plt.title('Agglomerative Linkage')
plt.xlabel('sepal')
plt.ylabel('distance')
plt.show()

plt.figure(1, figsize=(14,7))
dendrogram(hr_complete)
plt.title('Agglomerative Complete')
plt.xlabel('petal')
plt.ylabel('distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
aglo_linkage = AgglomerativeClustering(n_clusters= 3, affinity='euclidean', linkage='ward')
aglo_complete = AgglomerativeClustering(n_clusters= 5, affinity='euclidean')
```

```python
pr_aglo_linkage = aglo_linkage.fit_predict(points_sepal)
pr_aglo_complete = aglo_complete.fit_predict(points_petal)
```

```python
plt.figure(1, figsize=(8,8))
plt.scatter(points_sepal[pr_aglo_linkage==0,0], points_sepal[pr_aglo_linkage==0,1], 
            s=30, c='red', label='Cluster 1')
plt.scatter(points_sepal[pr_aglo_linkage==1,0], points_sepal[pr_aglo_linkage==1,1], 
            s=30, c='yellow', label='Cluster 2')
plt.scatter(points_sepal[pr_aglo_linkage==2,0], points_sepal[pr_aglo_linkage==2,1], 
            s=30, c='green', label='Cluster 3')
plt.title("Aglo Linkage")
plt.xlabel('Sepal length')
plt.ylabel('Sepal wdith')
plt.show()

plt.figure(1, figsize=(8,8))
plt.scatter(points_petal[pr_aglo_complete==0,0], points_petal[pr_aglo_complete==0,1], 
            s=30, c='red', label='Cluster 1')
plt.scatter(points_petal[pr_aglo_complete==1,0], points_petal[pr_aglo_complete==1,1], 
            s=30, c='blue', label='Cluster 2')
plt.scatter(points_petal[pr_aglo_complete==2,0], points_petal[pr_aglo_complete==2,1], 
            s=30, c='yellow', label='Cluster 3')
plt.scatter(points_petal[pr_aglo_complete==3,0], points_petal[pr_aglo_complete==3,1], 
            s=30, c='purple', label='Cluster 4')
plt.scatter(points_petal[pr_aglo_complete==4,0], points_petal[pr_aglo_complete==4,1], 
            s=30, c='green', label='Cluster 5')
plt.title("Aglo complete")
plt.xlabel('petal length')
plt.ylabel('petal wdith')
plt.show()
```
